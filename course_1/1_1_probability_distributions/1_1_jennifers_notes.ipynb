### Assigning probabilities

```python
# sample words and probabilities
candidate_words = ["star", "beef", "bottle"]

# The probabilities for each word are defined in the following list.
# They must add up to 1 (this becomes important when we get to embeddings).
your_mental_model = [0.99, 0.001, 0.009]

# This function runs several checks to verify that your_mental_model defines
# a proper probability distribution.
probabilities.test_probabilities(candidate_words, your_mental_model)

# Note: random.choices() is a library that uses a probability distribution
chosen_word = random.choices(candidate_words, weights=your_mental_model)
print(f"Twinkle twinkle little {chosen_word[0]}.")
```

### Task 1: Enter probabilities for each candidate word for the variable your_mental_model.
Note that
- your probabilities must add up to 1,
- are all > 0,
- every word has a non-zero probability.

Prompt: "Jide was **hungry** so she went looking for."

```python
candidate_words = [
    "food",
    "snacks",
    "leftovers",
    "her",
    "for",
    "water",
    "photosynthesis",
    "pyramid",
]

# Probabilities for candidate words
# prompt "Jide was hungry so she went looking for".
your_mental_model = [.30, .048, .30, .30, .001, .049, .001, .001 ]

# Checks to verify your_mental_model defines proper probability distribution.
probabilities.test_probabilities(candidate_words, your_mental_model)
```

```python
# Test
chosen_word = random.choices(candidate_words, weights=your_mental_model)

print(f"Jide was hungry so she went looking for {chosen_word[0]} ...")
```
