### Manually assign probabilities

```python
# sample words and probabilities
candidate_words = ["star", "beef", "bottle"]

# The probabilities for the above words are defined in the following list.
# They must add up to 1 (this becomes important when we get to embeddings).
my_probabilities = [0.99, 0.001, 0.009]

# Checks to verify that your_mental_model defines
# a proper probability distribution.
probabilities.test_probabilities(candidate_words, my_probabilities)

# Note: random.choices() is a library that uses a probability distribution
chosen_word = random.choices(candidate_words, weights=my_probabilities)
print(f"Twinkle twinkle little {chosen_word[0]}.")
```

### Manually assign probabilities for each candidate word for probability list.
Note that
- Probabilities must add up to 1,
- Are all > 0,

Prompt: "Jide was **hungry** so she went looking for."

```python
candidate_words = [
    "food",
    "snacks",
    "leftovers",
    "her",
    "for",
    "water",
    "photosynthesis",
    "pyramid",
]

# Probabilities for candidate words
# prompt "Jide was hungry so she went looking for".
hunger_probabilities = [.30, .048, .30, .30, .001, .049, .001, .001 ]

# Checks to verify your_mental_model defines proper probability distribution.
probabilities.test_probabilities(candidate_words, hunger_probabilities)

# Test
chosen_word = random.choices(candidate_words, weights=hunger_probabilities)
print(f"Jide was hungry so she went looking for {chosen_word[0]} ...")
```

### Context change

Change the prompt: "Jide was **thirsty** so she went looking for."

```python
candidate_words = [
    "food",
    "snacks",
    "leftovers",
    "her",
    "for",
    "water",
    "photosynthesis",
    "pyramid",
]
# New probabilities for when the context changes
thirst_probabilities = [.01, .01, .01, .34, .01, .6, .01, .01]

# checks
probabilities.test_probabilities(candidate_words, thirst_probabilities)

# Test
chosen_word = random.choices(candidate_words, weights=thirst_probabilities)
print(f"Jide was thirsty so she went looking for {chosen_word[0]}...")
```
